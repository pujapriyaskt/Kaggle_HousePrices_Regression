```python
x_train = df_train.drop(['SalePrice'],axis=1)
y_train = df_train['SalePrice']
```

```python
n_estimators = [100,500,900,1100,1500]
max_depth = [2,3,5,10,15]
booster = ['gbtree','gblinear']
learning_rate = [0.05,0.1,0.15,0.20]
min_cild_weight=[1,2,3,4]
base_score = [0.5]
```

```python
hyperparameter_grid = {'n_estimators':n_estimators,
                      'max_depth':max_depth,
                      'learning_rate': learning_rate,
                       'min_cild_weight': min_cild_weight,
                      'booster':booster,
                      'base_score':base_score
                      }
```

```python
import xgboost
Regressor= xgboost.XGBRegressor()
```

```python
from sklearn.model_selection import RandomizedSearchCV
random_cv = RandomizedSearchCV(estimator=Regressor,
                              param_distributions=hyperparameter_grid,
                              cv=5,n_iter=50,
                              scoring='neg_mean_absolute_error',n_jobs = 4,
                              verbose=5,
                              return_train_score=True,
                              random_state=42)
```

```python
random_cv.fit(x_train,y_train)
```

```python
random_cv.best_estimator_
```

```python
import xgboost
classifier= xgboost.XGBRegressor(base_score=0.5, booster='gbtree', colsample_bylevel=1,
       colsample_bynode=1, colsample_bytree=1, gamma=0,
       importance_type='gain', learning_rate=0.05, max_delta_step=0,
       max_depth=3, min_child_weight=1, min_cild_weight=1, missing=None,
       n_estimators=1100, n_jobs=1, nthread=None, objective='reg:linear',
       random_state=0, reg_alpha=0, reg_lambda=1, scale_pos_weight=1,
       seed=None, silent=None, subsample=1, verbosity=1)
classifier.fit(x_train,y_train)
```
